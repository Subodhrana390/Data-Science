```python
import pandas as pd
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# Load the dataset
df = pd.read_csv("student.csv")
X = df[['Hours']]
y = df[['Score']]

# Train the model
model = LinearRegression()
model.fit(X, y)

# Predict
predicted_score = model.predict(X)

# Evaluate the model
mae = mean_absolute_error(y, predicted_score) 
mse = mean_squared_error(y, predicted_score) 
rmse = np.sqrt(mse)

# Output evaluation results
print(f"MAE: {mae}")
print(f"MSE: {mse}")
print(f"RMSE: {rmse}")

# Predict for new input
new_hour = float(input("Enter hours studied: "))
new_pred = model.predict([[new_hour]])

print(f"Predicted Score: {new_pred[0][0]:.2f}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Sample Data: Weekly Study Hours and Final Exam Scores
data = {
    'Study_Hours': [2, 5, 7, 3, 8, 12, 10, 4, 6, 9],
    'Exam_Scores': [50, 60, 65, 55, 70, 90, 85, 58, 68, 78]
}

# Create DataFrame
df = pd.DataFrame(data)

# Define X (independent variable) and y (dependent variable)
X = df[['Study_Hours']]
y = df['Exam_Scores']

# Create and train linear regression model
model = LinearRegression()
model.fit(X, y)

# Make predictions
df['Predicted_Scores'] = model.predict(X)

# Print model coefficients
print("Intercept (β₀):", model.intercept_)
print("Slope (β₁):", model.coef_[0])

# Calculate R-squared
r2 = r2_score(y, df['Predicted_Scores'])
print("R-squared:", round(r2, 3))

# Plot the results
plt.figure(figsize=(8, 6))
plt.scatter(df['Study_Hours'], df['Exam_Scores'], color='blue', label='Actual Scores')
plt.plot(df['Study_Hours'], df['Predicted_Scores'], color='red', label='Regression Line')
plt.title('Study Hours vs Final Exam Scores')
plt.xlabel('Weekly Study Hours')
plt.ylabel('Final Exam Scores')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
