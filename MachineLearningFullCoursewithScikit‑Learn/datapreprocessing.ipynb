```python
import pandas as pd

data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [25, None, 35],
    "City": ["New York", "Los Angeles", None],
    "Salary": [70000, None, 80000]
}

df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)

print("\nDataFrame with NaN values:")
print(pd.isnull(df).sum())
# df_drop = df.dropna()
# print("\nDataFrame after dropping rows with any NaN values:")
# print(df_drop)
print("\nDataFrame after filling NaN values:")

df.fillna({
    'Age': df['Age'].mean(),
    'Salary': df['Salary'].mean(),
    'City': df['City'].mode()[1]
}, inplace=True)
print(df)
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from random import randint

data = {
    "study_hour": [randint(1, 10) for _ in range(30)],  
    "score": [randint(20, 100) for _ in range(30)],      
}

df = pd.DataFrame(data)
print(df.head())

#standard Scalar
standard_scalar = StandardScaler()
standard_scaled = standard_scalar.fit_transform(df)

print("Standard Scalar Output")
print(pd.DataFrame(standard_scaled,columns=["study_hour","score"]).head())

#standard Scalar
minMax_scalar = MinMaxScaler()
minMax_scaled = minMax_scalar.fit_transform(df)

print("minMax Scalar Output")
print(pd.DataFrame(minMax_scaled,columns=["study_hour","score"]).head())

x = df[["study_hour"]]   
y = df[["score"]]       

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

print("Training Data")
print(X_train)

print("Testing Data")
print(X_test)
```
