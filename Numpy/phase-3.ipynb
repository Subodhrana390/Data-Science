# Advance Operation with Bussiness Examples

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 115000, 180000, 220000, 250000],  # Beijing Bites
    [2, 120000, 140000, 160000, 190000],   # Chai Point
    [3, 200000, 230000, 260000, 300000],  # Paradise Biryani
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]  # Pizza Hub
])

print("Sales Data shape :\n", sales_data.shape)
print("Sales Data for first 3 Restaurant :\n", sales_data[:,1:])
```

```python
sum_sales = np.sum(sales_data[:, 1:], axis=0)
print("Total Sales from 2021 to 2024 :\n", sum_sales)

print("Per year sales growth :\n", np.diff(sum_sales))

print("Minimum sales in a year : ", np.min(sales_data[:,1:],axis=1))
print("Maximum sales in a year : ", np.max(sales_data[:,1:],axis=0))
print("Average sales in a year : ", np.mean(sales_data[:,1:],axis=1))

# Visualization
years = [2021, 2022, 2023, 2024] 
plt.plot(years, sum_sales, marker='o') 
```

# Culmative Sales 

```python
cumulative_sales = np.cumsum(sales_data[:, 1:], axis=1)
print("Cumulative Sales from 2021 to 2024 :\n", cumulative_sales)

plt.figure(figsize=(8, 6))
plt.plot(np.mean(cumulative_sales, axis=0), marker='o', color='orange')
plt.title('Average Cumulative Sales Over Years')
plt.xlabel('Years')
plt.ylabel('Average Cumulative Sales')
plt.grid(True)
plt.show()
```

```python
vector_1 = np.array([1, 2, 3]) 
vector_2 = np.array([4, 5, 6])
vector_addition = np.add(vector_1, vector_2)
print("addition Product of vector_1 and vector_2 : ", vector_addition)
print("Multplication Product of vector_1 and vector_2 : ", np.multiply(vector_1, vector_2))
print("Dot Product of vector_1 and vector_2 : ", np.dot(vector_1, vector_2))

angle = np.arccos(np.dot(vector_1, vector_2) / (np.linalg.norm(vector_1) * np.linalg.norm(vector_2)))
print("Angle between vector_1 and vector_2 in radians : ", angle)
```
