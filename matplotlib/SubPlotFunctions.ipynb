```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 15, 25]

fig, ax = plt.subplots(1, 2, figsize=(10, 5))

# First subplot - Line plot
ax[0].plot(x, y, marker="o",color="gray")
ax[0].set_title("Line Plot")

# Second subplot - Scatter plot
ax[1].bar(x, y, color="red")
ax[1].set_title("Scatter Plot")

fig.suptitle("Comparsion of Line and bar Chart")

plt.tight_layout()

plt.show()
```

```python
#SaveFig - Save the current figure
# Syntax - SaveFig(filename.extension,dpi=value,bbox_inches='tight')

import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 15, 25]

fig, ax = plt.subplots(1, 2, figsize=(10, 5))

# First subplot - Line plot
ax[0].plot(x, y, marker="o",color="gray")
ax[0].set_title("Line Plot")

# Second subplot - Scatter plot
ax[1].bar(x, y, color="red")
ax[1].set_title("Scatter Plot")

fig.suptitle("Comparsion of Line and bar Chart")

plt.tight_layout()

plt.savefig("Line_Bar_Chart.png",dpi=300,bbox_inches='tight')
plt.show()
```
