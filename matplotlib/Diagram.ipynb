# BarChart Visualization

```python
product = ['A', 'B', 'C', 'D']
sales = [120, 90, 150, 70]
plt.barh(product, sales, color='skyblue',label="Sales 2025")
plt.xlabel('Product')
plt.ylabel('Sales')
plt.title('Product Sales')
plt.legend(loc='upper right')
plt.show()
```

```python
# Data
companies = ['Company A', 'Company B', 'Company C', 'Company D']
sales = [350, 250, 150, 100]

# Create pie chart
plt.figure(figsize=(4, 4))  # optional: makes chart bigger
plt.pie(sales, labels=companies, autopct='%1.1f%%', startangle=140)

# Add title
plt.title('Market Share of Companies Selling Products Across India')

# Equal aspect ratio ensures pie is a circle

plt.axis('equal')   

# Show chart
plt.show()
```

```python
# Student marks
marks = [45, 67, 89, 70, 56, 78, 82, 61, 55, 90,
         73, 69, 84, 76, 50, 62, 59, 88, 95, 80]

# Create histogram
plt.hist(marks, bins=5, color='gray', edgecolor='black')

# Add labels and title
plt.xlabel('Marks Range')
plt.ylabel('Number of Students')
plt.title('Distribution of Student Marks')

# Show plot
plt.show()
```

Scatter Plot

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
hours = [1, 2, 2.5, 3, 3.5, 4, 4.5, 5, 5.5, 6, 6.5, 7, 7.5, 8, 9]
scores = [35, 40, 45, 50, 52, 55, 60, 65, 68, 70, 72, 78, 80, 85, 90]

# Trendline using numpy (linear regression)
coeffs = np.polyfit(hours, scores, deg=1)
trendline = np.poly1d(coeffs)
line_x = np.linspace(min(hours), max(hours), 100)
line_y = trendline(line_x)

# Plot
plt.figure(figsize=(10, 6))
scatter = plt.scatter(hours, scores, 
                      c=scores, cmap='plasma', 
                      s=100, edgecolors='black', alpha=0.9)

# Trendline
plt.plot(line_x, line_y, color='black', linestyle='--', linewidth=2, label='Trendline')

# Labels
plt.title(' Study Hours vs Exam Score', fontsize=16, fontweight='bold')
plt.xlabel('Hours Studied', fontsize=12)
plt.ylabel('Score', fontsize=12)

# Grid, style, legend
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.colorbar(scatter, label='Score')

# Optional: Add data labels
for h, s in zip(hours, scores):
    plt.text(h, s + 1, f'{s}', fontsize=9, ha='center', va='bottom')

# Show plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Class A
hours_a = [2, 3, 4, 5, 6, 7]
scores_a = [40, 50, 55, 65, 70, 75]

# Class B
hours_b = [2.5, 3.5, 4.5, 5.5, 6.5, 7.5]
scores_b = [45, 52, 60, 68, 74, 80]

# Plot
plt.figure(figsize=(10, 6))

# Scatter plots for each class
plt.scatter(hours_a, scores_a, color='blue', marker='o', s=100, label='Class A', edgecolor='black')
plt.scatter(hours_b, scores_b, color='green', marker='^', s=100, label='Class B', edgecolor='black')

# Labels and title
plt.title(' Comparison of Class A and B: Study Hours vs Exam Scores', fontsize=14, fontweight='bold')
plt.xlabel('Hours Studied')
plt.ylabel('Score')

# Grid and legend
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.tight_layout()

# Show plot
plt.show()
```
