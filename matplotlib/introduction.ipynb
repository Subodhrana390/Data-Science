# plot functions 

```python
x = [1, 2, 3, 4, 5, 6] 
y = [12, 45, 23, 56, 74, 74] 

plt.plot(x, y, color="blue", linestyle="--", linewidth=2,marker='o',label='Monthly Sales')  
plt.title('Monthly Sales')
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend(loc='upper left',fontsize=12)
plt.grid(color='gray',linestyle=':',linewidth=1)
plt.xticks(x, ['January', 'February', 'March', 'April', 'May', 'June'])
plt.show()
```
